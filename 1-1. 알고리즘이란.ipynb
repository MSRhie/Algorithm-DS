# 문제 : 입력받는 세 정수의 최댓값 구하기

```python
# 1. 순서도 보고 문제 풀기 24.11.05

a = input()
b = input()
c = input()

maximum = a

if b > maximum :
  maximum = b
if c > maximum :
  maximum = c

print('최댓값 : ', maximum)
```

```python
# 2. 함수 구현  24.11.05
def max3(a,b,c) :
  '''a,b,c의 최댓값을 구하여 반환'''
  maximum2 = a
  if b > maximum2 :
    maximum2 = b
  if c > maximum2 :
    maximum2 = c
  return maximum2

print(max3(4,5,6))
```

# 문제 : 입력받는 세 정수의 대소 관계로 중앙값 구하기

```python
# 1. 순서도 보고 코딩

def med(a,b,c) :
  if a >= b :
    if b >= c :
      return b
  if b >= c :
    if c >= a :
      return c
  if c >= a :
    if a >= b :
      return a
  else :
    return b

print(med(1,2,3))
```
