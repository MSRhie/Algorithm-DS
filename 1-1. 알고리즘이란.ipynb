# 문제 : 입력받는 세 정수의 최댓값 구하기

```python
# 1. 순서도 보고 문제 풀기 24.11.05

a = input()
b = input()
c = input()

maximum = a

if b > maximum :
  maximum = b
if c > maximum :
  maximum = c

print('최댓값 : ', maximum)
```

```python
# 2. 함수 구현  24.11.05
def max3(a,b,c) :
  '''a,b,c의 최댓값을 구하여 반환'''
  maximum2 = a
  if b > maximum2 :
    maximum2 = b
  if c > maximum2 :
    maximum2 = c
  return maximum2

print(max3(4,5,6))
```

# 문제 : 입력받는 세 정수의 대소 관계로 중앙값 구하기

```python
# 1. 순서도 보고 코딩

def med(a,b,c) :
  if b <= a :
    if c <= b :
      return b
    elif c <= a :
      return a
    else :
      return c
  elif c <= a :
    return a
  elif b <= a :
    return b
  else :
    return c

print(med(4,2,3))
```

# 문제 : 입력받은 정숫값의 부호(양수, 음수, 0)를 판단하여 출력해라

```python
# 1.순서도 작성
# 2.순서도 보고 코딩

def pm0(n) :
  if n > 0 :
    print("양수")
  elif n < 0 :
    print("음수")
  else :
    print("0")

pm0(-1)
```

# 문제 : 1부터 n까지 정수의 합 구해라

```python
# 방법 1 : while 문으로 작성
# 1.순서도 작성
# 2.순서도 보고 코딩

def while_sum(n) :
  sum = 0
  i = 1
  while i <= n :
    sum += i
    i += 1
  return sum

while_sum(10)

# 방법 2 : for 문으로 작성
# 카운트용 변수 i를 제외하고 변수가 하나 일땐 While 보다 for 가 더 좋다.
# 1.순서도 작성
# 2.순서도 보고 코딩

def for_sum(n) :
  sum = 0
  for i in range(1,n+1) :
    sum = sum + i
  return sum
for_sum(100)

# 방법 3 : 공식으로 처리
def formula(n) :
  return n * (n+1)//2

formula(100)
```

# 문제 : 임의의 두 값이 주어졌을 때 연속하는 정수의 합을 구하기 위해 값 정렬하기

## 시작하는 값이 1이 아닌 임의의 두 정수를 입력받았을 때 합계 출력하기

```python
def two_sum(a,b) :
  sum = 0

  if a > b :
    a, b = b, a

  for i in range(a,b+1) :
    sum += i
  return sum

two_sum(1,10)
```

## 정수의 합을 구하는 과정과 최종값 출력하기

```python
def two_sum(a,b) :
  sum = 0

  if a > b :
    a, b = b, a

  for i in range(a,b+1) :
    if i < b :
      print(f'{i} + ', end = '') # end = '' 가 있으면 줄바꿈을 안한다.
    else :
      print(f'{i} = ', end = '')
    sum += i
  return sum

two_sum(1,10)
```

* 만약 10,000개의 데이터를 더한다하면, for문 안의 if는 9,999번 True를, 1번의 False를 출력할 것이다. 이경우 for문안에 if를 넣기보다 마지막 값을 for문에서 꺼내어 작성하는게 바람직한 코드이다.

```python
def two_sum1(a,b) :
  sum = 0

  if a > b :
    a, b = b, a

  for i in range(a,b) :
    print(f'{i} + ', end = '')
    sum += i
  print(f'{b} = ', end = '')
  sum += b
  return sum

two_sum1(1,10)
```

# 문제 : 입력받은 숫자만큼 줄바꿈없이 +,-를 반복하여 연속해서 출력해라

* 예시 : +-+-+-...

```python
def repeat_pm(n):
  for i in range(1,n+1) :
    if i % 2 :
      print('+', end='')
    else :
      print('-', end='')

repeat_pm(5)
```

* 이경우 두가지 문제가 있다.
1. 카운트용 변수의 시작값이 바뀌면 유연하게 대처가 어렵다. (홀수로 시작하냐, 짝수로 시작하냐에 따라 처음 문자값이 바뀐다.)
2. if문이 for안에 들어가 있어 반복할때마다 조건문을 실행한다.

```python
def repeat_pm1(n) :
  for _ in range(n//2) :
    print('+-', end='')

  if n % 2 :
    print('+', end='')

  print()

repeat_pm1(5)
```
