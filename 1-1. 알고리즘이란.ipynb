# 문제 : 입력받는 세 정수의 최댓값 구하기

```python
# 1. 순서도 보고 문제 풀기 24.11.05

a = input()
b = input()
c = input()

maximum = a

if b > maximum :
  maximum = b
if c > maximum :
  maximum = c

print('최댓값 : ', maximum)
```

```python
# 2. 함수 구현  24.11.05
def max3(a,b,c) :
  '''a,b,c의 최댓값을 구하여 반환'''
  maximum2 = a
  if b > maximum2 :
    maximum2 = b
  if c > maximum2 :
    maximum2 = c
  return maximum2

print(max3(4,5,6))
```

# 문제 : 입력받는 세 정수의 대소 관계로 중앙값 구하기

```python
# 1. 순서도 보고 코딩

def med(a,b,c) :
  if b <= a :
    if c <= b :
      return b
    elif c <= a :
      return a
    else :
      return c
  elif c <= a :
    return a
  elif b <= a :
    return b
  else :
    return c

print(med(4,2,3))
```

# 문제 : 1부터 n까지 정수의 합 구해라

```python
# 방법 1 : while 문으로 작성
# 1.순서도 작성
# 2.순서도 보고 코딩

def while_sum(n) :
  sum = 0
  i = 1
  while i <= n :
    sum += i
    i += 1
  return sum

while_sum(10)

# 방법 2 : for 문으로 작성
# 카운트용 변수 i를 제외하고 변수가 하나 일땐 While 보다 for 가 더 좋다.
# 1.순서도 작성
# 2.순서도 보고 코딩

def for_sum(n) :
  sum = 0
  for i in range(1,n+1) :
    sum = sum + i
  return sum
for_sum(100)

# 방법 3 : 공식으로 처리
def formula(n) :
  return n * (n+1)//2

formula(100)
```

# 문제 : 임의의 두 값이 주어졌을 때 연속하는 정수의 합을 구하기 위해 값 정렬하기

## 임의의 두 정수를 입력받았을 때 합계 출력하기

```python
def two_sum(a,b) :
  sum = 0

  if a > b :
    a, b = b, a

  for i in range(a,b+1) :
    sum += i
  return sum

two_sum(1,10)
```

## 정수의 합을 구하는 과정과 최종값 출력하기

```python
def two_sum2(a,b) :
  sum = 0

  if a > b :
    a, b = b, a

  for i in range(a,b+1) :
    if i < b :
      print(f'{i} + ', end = '') # end = '' 가 있으면 줄바꿈을 안한다.
    else :
      print(f'{i} = ', end = '')
    sum += i
  return sum

two_sum2(1,10)
```

* 만약 10,000개의 데이터를 더한다하면, for문 안의 if는 9,999번 True를, 1번의 False를 출력할 것이다. 이경우 for문안에 if를 넣기보다 마지막 값을 for문에서 꺼내어 작성하는게 바람직한 코드이다.

```python
def two_sum3(a,b) :
  sum = 0

  if a > b :
    a, b = b, a

  for i in range(a,b) :
    print(f'{i} + ', end = '')
    sum += i
  print(f'{b} = ', end = '')
  sum += b
  return sum

two_sum3(1,10)
```

# 문제 : 입력받은 숫자만큼 줄바꿈없이 +,-를 반복하여 연속해서 출력해라

* 예시 : +-+-+-...

```python
def repeat_pm(n):
  for i in range(1,n+1) :
    if i % 2 :
      print('+', end='')
    else :
      print('-', end='')

repeat_pm(5)
```

* 이경우 두가지 문제문제를 아래 코드로 해결이 가능하다.
1. 카운트용 변수 i의 시작값이 바뀌면 유연하게 대처가 어렵다. (홀수로 시작하냐, 짝수로 시작하냐에 따라 처음 문자값이 바뀐다.)
 * 카운트용 변수 i를 없애고 루프만 돌린다. 따라서 값 시작시점이 바꾸면 range(시작시점값, n//2+시작시점값)을 넣으면 대처가 가능해졌다.
2. if문이 for안에 들어가 있어 반복할때마다 조건문을 실행한다.
 * range(n//2)로 몫의 갯수만큼 +-를 반복(짝수 문제해결) 하고 2로 나눴을 때 나머지가있다면 마지막에 +를 더 추가하여 홀, 짝 문제를 해결했다.

```python
def repeat_pm1(n) :
  for _ in range(n//2) :
    print('+-', end='')

  if n % 2 :
    print('+', end='')

  print()

repeat_pm1(5)
```

# [못품]문제 : *를 입력한 갯수(n개)만큼 출력하고 입력한 갯수(w)마다 줄바꿈해라

* 줄바꿈은 print() 사용

```python
def star_enter(n,w) :
  for i in range(n) :
    print('*', end='')
    if i % w == w - 1 :
      print()
  if n % w : # 못푼 이유 : 이 if문은 왜 필요하는가? 뺴도 같은 값 나옴
    print()

star_enter(19,5)
```

* for 내 if문 반복 제거

```python
def star_enter1(n,w) :
  for _ in range(n//w): # //는 몫구함
    print('*' * w)
  rest = n % w
  if rest :
    print('*' * rest)

star_enter1(19,5)
```

# 문제 : 길이와 넓이가 모두 정수인 직사각형의 넓이로 변의 길이 구하기

* 이때 짧은 변과 긴변은 구분하지 않는다. 만약 16 x 2이면 2 x 16 두개가 아닌 하나만 출력해야 한다.

```python
def space(area) :
  for i in range(1, area + 1) :
    if i * i > area : break # 짧은 변 긴변 구별 안함
    if area % i : continue # 나누었을 때 나머지가 나온다면 아래 조건식을 건너뛰고 다시 for문 첫 루틴으로 돌아감.
    print(f'{i} x {area // i}')

space(32)
```

# 10~99 사이의 난수 n개 생성하기(단, 13이 나오면 중단)

```python
import random

def random_num(n) :
  for _ in range(n) :
    r = random.randint(10,99)
    print(r, end=' ')
    if r == 13 :
      print('\n프로그램을 종료합니다.')
      break
  else : # While문이나 for문과 같은 반복문의 끝부분에 else를 둘 수 있다.
    print('\n 난수 생성을 종료합니다.')

random_num(2)
```

# 문제 : 1부터 12까지 8을 건너뛰고 출력해라

```python
def jump(n) :
  for i in range(1,n+1) :
    if i == 8 :
      continue # 앞에서 배운 coutiune 활용
    else :
      print(f'{i}', end=' ')

jump(12)
```

# 문제 : 1~9단까지의 구구단 곱셈표를 출력해라

```python
def gugudan(n) :
  for i in range(1,n+1) :
    print()
    for j in range(1,n+1) :
      ij = i * j
      print(f'{ij}', end = ' ')

def gugudan1(n) :
  for i in range(1,n+1) :
    for j in range(1,n+1) :
      print(f'{i*j:3}', end = '') # 열간격 주는 옵션 :3
    print()

gugudan1(9)
```
