# 문제 : 입력받는 세 정수의 최댓값 구하기

```python
# 1. 순서도 보고 문제 풀기 24.11.05

a = input()
b = input()
c = input()

maximum = a

if b > maximum :
  maximum = b
if c > maximum :
  maximum = c

print('최댓값 : ', maximum)
```

```python
# 2. 함수 구현  24.11.05
def max3(a,b,c) :
  '''a,b,c의 최댓값을 구하여 반환'''
  maximum2 = a
  if b > maximum2 :
    maximum2 = b
  if c > maximum2 :
    maximum2 = c
  return maximum2

print(max3(4,5,6))
```

# 문제 : 입력받는 세 정수의 대소 관계로 중앙값 구하기

```python
# 1. 순서도 보고 코딩

def med(a,b,c) :
  if b <= a :
    if c <= b :
      return b
    elif c <= a :
      return a
    else :
      return c
  elif c <= a :
    return a
  elif b <= a :
    return b
  else :
    return c

print(med(4,2,3))
```

# 문제 : 입력받은 정숫값의 부호(양수, 음수, 0)를 판단하여 출력해라

```python
# 1.순서도 작성
# 2.순서도 보고 코딩

def pm0(n) :
  if n > 0 :
    print("양수")
  elif n < 0 :
    print("음수")
  else :
    print("0")

pm0(-1)
```

# 문제 : 1부터 n까지 정수의 합 구해라

```python
# 방법 1 : while 문으로 작성
# 1.순서도 작성
# 2.순서도 보고 코딩

def while_sum(n) :
  sum = 0
  i = 1
  while i <= n :
    sum += i
    i += 1
  return sum

while_sum(10)

# 방법 2 : for 문으로 작성
# 카운트용 변수 i를 제외하고 변수가 하나 일땐 While 보다 for 가 더 좋다.
# 1.순서도 작성
# 2.순서도 보고 코딩

def for_sum(n) :
  sum = 0
  for i in range(1,n+1) :
    sum = sum + i
  return sum
for_sum(100)

# 방법 3 : 공식으로 처리
def formula(n) :
  return n * (n+1)//2

formula(100)
```
