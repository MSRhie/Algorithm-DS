# 문제 : 어떤 학생 그룹의 시험 점수를 집계하여라

```python
def student() :
  score1 = int(input('1번 학생의 점수를 입력하시오.'))
  score2 = int(input('2번 학생의 점수를 입력하시오.'))
  score3 = int(input('3번 학생의 점수를 입력하시오.'))
  score4 = int(input('4번 학생의 점수를 입력하시오.'))
  score5 = int(input('5번 학생의 점수를 입력하시오.'))

  total = 0
  total += score1
  total += score2
  total += score3
  total += score4
  total += score5

  print(f'합계는 {total}점입니다.')
  print(f'평균은 {total/5}점입니다.')
```

```python
student()
```

* 1) 학생추가하려면 score6 = int(input('5번 학생의 점수를 입력하시오.')) 와 total+=score6 를 또 입력해야 함
* 2) 이미 입력받은 값을 확인하거나 변경하려면 각 기능별로 프로그래밍을 또 해야함 (if, for문 써가며..)

## 슬라이스식으로 원소에 접근하기

```python
x = [1,2,3,4,5]
x[:] # 전체 출력 [1,2,3,4,5]
x[0:6] # x x[:]와 같음 전체 출력 [1,2,3,4,5]

x[:3] # 맨 앞에서 3개 출력 [1, 2, 3]
x[1:] # 두번째자리부터 끝까지 출력 [2, 3, 4, 5]
x[:-1] # 맨 앞에서 끝에서 두번째 자리부터 출력

x[0:6:2] # 첫번째 자리 출력후 +2 자리마다 출력 (홀수 출력) [1,3,5]
x[::2] # 위와 같음 (홀수 출력)

x[1:6:2] # 두번째 자리 출력후 +2 자리마다 출력 (짝수 출력) # for, if 문 안짜도 바로 홀/짝수 출력 가능 [2,4]
x[1::1] # 위와 같음 (짝수 출력)

x[-3:-1] # 뒤에서 3번째부터 뒤에서 1번째전까지 출력 [3,4]
x[-3:] # 뒤에서 1번째 값부터 마지막값까지 출력 [3,4,5]
x[::-1] # 뒤에서부터 전부 출력 [5, 4, 3, 2, 1]

x[:7] # x[i:j:k] 에서 i, j가 값을 벗어나면 자동으로 len(x)값이 7자리에 들어가서 오류발생하지 않음
```
